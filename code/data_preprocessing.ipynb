```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
seoul = pd.read_csv('./서울시 코로나19 확진자 현황.csv',engine='python')
```

```python
seoul = seoul[['연번', '확진일', '지역','노출여부']]
```

```python
seoul
```

```python
seoul.isna().sum()
```

```python
time1 = datetime(2020, 1, 24).date()
list1 = [str(time1)]
for i in range(1,625):
    time2 = time1 + timedelta(days = i)
    list1.append(str(time2))
```

```python
list1
```

```python
seoul_count = pd.DataFrame(list1,columns = ['확진일'])
```

```python
seoul_count
```

```python
col = seoul['지역'].unique() 
```

```python
for i in col:
    a = seoul[seoul['지역'] == i].groupby('확진일').count()
    a.reset_index(inplace = True)
    a = a[['확진일','연번']]
    a.rename(columns = {'연번' : i}, inplace = True)
    seoul_count = pd.merge(seoul_count, a, how = "left",on = '확진일')
```

```python
seoul_count.fillna(0,inplace = True)
```

```python
seoul_count.columns
```

```python
seoul_count.drop(['타시도','기타','확진일'],axis = 1, inplace = True)
```

```python
seoul_count
```

```python
seoul_count['서울시전체'] = seoul_count.sum(axis = 1)
```

```python
seoul_count
```

```python
seoul_count.to_csv('seoul_count.csv',index = False,encoding = 'utf-8-sig')
```
